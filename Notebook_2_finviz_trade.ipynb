***NOTEBOOK 2 : EXTRACTING DATA FROM THE PICKLE FILE TO CREATE AN UPDATED TRADE CSV FILE***

**STEP 2.1: IMPORTING AND LOADING THE PICKLE FILE**

```python
import pickle
with open("C:/Users/liemu/Downloads/Finviz/Sectors/finviz.pickle", "rb") as f:
    data = pickle.load(f)
```

```python
data['AA']
```

Reviewing the nature of "trade" data format in the file with the TSLA key

```python
for key in data['TSLA']['trade']:
    print(key)
```

**STEP 2.2 : GETTING THE TRADE DATA IN A LIST FORMAT**

```python
# Create an empty list
trade_ls = []
# Iterating through all ticker symbols in the data dictionary
for key in data.keys():
    # Check if the ticker symbol has a value for "trade" in the data dictionary, and proceed if it does else skip
    if 'trade' in data[key].keys():
        # Iterating through all the rows of "trade" for each "ticker" in the data dictionary
        for row in data[key]['trade']:
            # Appending the respective values to the trade_ls list
            trade_ls.append(row)
```

```python
# Checking the length of trade datas in trade_ls
len(trade_ls)
```

```python
# Checking the first value in the trade_ls list
trade_ls[0]
```

**STEP 2.2.1 : CONVERTING THE TRADE_LS LIST DATA INTO A DATAFRAME FORMAT**

```python
import pandas as pd
df = pd.DataFrame(trade_ls, columns=['Date', 'Open', 'Close', 'Volume', 'Company', 'Industry', 'Sector', 'Country'])
```

**STEP 2.3 : CALCULATING THE RETURN VALUE FOR EACH TRADE DATA IN DF**

```python
df['Return'] = df['Close'] - df['Open']
df
```

**STEP 2.3.1 : CALCULATING THE RETURN PERCENTAGE VALUE FOR EACH TRADE DATA IN DF**

```python
df['Return_pct'] = (df['Return'] / df['Open']) * 100
df['Return_pct'] = df['Return_pct'].round(2)
df
```

**STEP 2.4 : GETTING THE SENTIMENT DATA 'comp' FROM DATA IN LIST FORMAT**

```python
# Create an empty list
comp_ls = []
# Iterating through all ticker symbols in the data dictionary
for key in data.keys():
    # Check if the ticker symbol has a value for "comp" in the data dictionary, and proceed if it does else skip
    if 'comp' in data[key].keys():
        # Iterating through all the rows of "comp" for each "ticker" in the data dictionary
        for row in data[key]['comp']:
            # Appending the respective values to the comp_ls list
            comp_ls.append(row)
```

```python
# Checking the length of trade datas in trade_ls
len(comp_ls)
```

```python
# Checking the first value in the comp_ls list
comp_ls[0]
```

**STEP 2.4.1 : CONVERTING THE COMP_LS LIST DATA INTO A DATAFRAME FORMAT**

```python
df1 = pd.DataFrame(comp_ls, columns=['Date', 'compound', 'neg', 'pos', 'neu', 'Company', 'Industry', 'Sector', 'Country'])
# Calculating the sentiment score
df1['sentiment'] = (df1['compound'] * 10).round()
# Converting the sentiment column to integer data type
df1['sentiment'] = df1['sentiment'].astype(int)
df1
```

**STEP 2.5 : MERGING THE TWO DATAFRAMES ON COMPANY AND DATE**

```python
merged_df = pd.merge(df, df1, on=["Company", "Date", "Industry", "Sector", "Country"])
# Setting index column as Date
merged_df = merged_df.set_index('Date')
merged_df
```

**STEP 2.6 : SAVING THE TRADE FILE IN CSV FORMAT FOR TRAINING COMPLEX MODELS**

```python
# Saving File as CSV
merged_df.to_csv('./updated_trade.csv')
```
